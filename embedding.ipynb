```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

```python
import tensorflow as tf 

from tensorflow.keras.preprocessing.text import one_hot
```

```python
Sent=['The glass of milk',
      'The glass of juice',
      'The cup of tea',
      'I am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'your videos are good']
```

```python
Sent
```

```python
# Define the vocubulary size

voc_size=10000
```

```python
# OneHot Representation

one_hot_repr=[one_hot(words, voc_size) for words in Sent]
one_hot_repr
```

```python
# Word Embedding Representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences # helps to make the similar length of words in each sentence

from tensorflow.keras.models import Sequential
```

```python
sent_length=8
embedded_doc=pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
```

```python
print(embedded_doc)
```

```python
pad_sequences(one_hot_repr, padding='post', maxlen=sent_length)
```

```python
# Feature Representation

dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size, dim, input_length=sent_length))
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```

```python
model.predict(embedded_doc[[0]])
```
